```python
import pandas as pd
```

```python
data = pd.read_json("gh-push-event.json", orient='records')
```

```python
data.head()
```

```python
data['month'] = (data['quarter'] * 3)
```

```python
data.sample(10)
```

```python
data['date'] = data['year'].astype('str') + "-" + data['month'].astype('str').str.zfill(2) + "-30"
```

```python
data.sample(10)
```

```python
data['value'] = data['count']
```

```python
data = data[['date', 'name', 'value']]
```
