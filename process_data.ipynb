```python
covid_19_raw = pd.read_csv('covid-19/time-series-19-covid-combined.csv')
covid_19 = pd.DataFrame()
covid_19['date'] = covid_19_raw['Date']
covid_19['name'] = covid_19_raw['Country/Region']
covid_19['value'] = covid_19_raw['Confirmed']
covid_19.sort_values(by="value", ascending=False, inplace=True)
covid_19.drop_duplicates(subset=['date', 'name'], inplace=True)
covid_19.to_csv('_processed/covid-19.csv', index=False)
covid_19.head()
```

```python
exchange_raw = pd.read_csv('exchange-rates/monthly.csv')
exchange = pd.DataFrame()
exchange['date'] = exchange_raw['Date']
exchange['name'] = exchange_raw['Country']
exchange['value'] = exchange_raw['Exchange rate'].round(2)
exchange.to_csv('_processed/exchange-rates.csv', index=False)
exchange.head()
```

```python
gdp_raw = pd.read_csv('gdp/gdp.csv')
gdp = pd.DataFrame()
gdp['date'] = gdp_raw['Year'].astype('str') + '-01-01'
gdp['name'] = gdp_raw['Country Name']
gdp['code'] = gdp_raw['Country Code']
gdp['value'] = gdp_raw['Value'].divide(10**9).round(2)
country_codes_raw = pd.read_csv('helper/country-codes.csv', keep_default_na=False, na_values=['_'])
country_codes = pd.DataFrame()
country_codes['code'] = country_codes_raw['ISO3166-1-Alpha-3']
country_codes['code2'] = country_codes_raw['ISO3166-1-Alpha-2']
country_codes['Continent'] = country_codes_raw['Continent']
country_codes.head()
gdp = gdp.merge(country_codes, on='code')
continent_codes = pd.read_csv('helper/continent-codes.csv', keep_default_na=False, na_values=['_'])
gdp = gdp.merge(continent_codes, left_on='Continent', right_on='Code')
gdp.head()
gdp_final = pd.DataFrame()
gdp_final['date'] = gdp['date']
gdp_final['name'] = gdp['name']
gdp_final['code'] = gdp['code2']
gdp_final['group'] = gdp['Name']
gdp_final['value'] = gdp['value']
gdp_final.to_csv('_processed/gdp.csv', index=False)
gdp_final.shape
gdp_final.head()
```

```python
inflation_raw = pd.read_csv('inflation-gdp/inflation-gdp.csv')
inflation = pd.DataFrame()
inflation['date'] = inflation_raw['Year'].astype('str') + '-01-01'
inflation['name'] = inflation_raw['Country']
inflation['code'] = inflation_raw['Country Code']
inflation['value'] = inflation_raw['Inflation'].round(2)

inflation = inflation.merge(country_codes, on='code')
inflation = inflation.merge(continent_codes, left_on='Continent', right_on='Code')

inflation_final = pd.DataFrame()
inflation_final['date'] = inflation['date']
inflation_final['name'] = inflation['name']
inflation_final['code'] = inflation['code2']
inflation_final['group'] = inflation['Name']
inflation_final['value'] = inflation['value']
inflation_final.to_csv('_processed/inflation.csv', index=False)
inflation_final.head()
```

```python
population_raw = pd.read_csv('population/population.csv')
population = pd.DataFrame()
population['date'] = population_raw['Year'].astype('str') + '-01-01'
population['name'] = population_raw['Country Name']
population['code'] = population_raw['Country Code']
population['value'] = population_raw['Value'].divide(10**6).round(2)

population = population.merge(country_codes, on='code')
population = population.merge(continent_codes, left_on='Continent', right_on='Code')

population_final = pd.DataFrame()
population_final['date'] = population['date']
population_final['name'] = population['name']
population_final['code'] = population['code2']
population_final['group'] = population['Name']
population_final['value'] = population['value']
population_final.to_csv('_processed/population.csv', index=False)
population_final.head()
```
