```python
data.drop(['color', 'lastValue'], axis=1, inplace=True)
```

```python
data['icon_url'] = 'https:' + data['img']
data['icon_file'] = data['name'].str.lower().str.replace(' ', '_') + '.svg.png'
data.sample(10)
```

```python
icons = data[['name', 'icon_url', 'icon_file']]
```

```python
icons.shape
```

```python
icons = icons.drop_duplicates()
```

```python
icons.shape
```

```python
import os
from urllib.parse import urlparse
import urllib.request as urllib2

def download_img(url, file):
    testfile = urllib2.URLopener()
    testfile.retrieve(url, filename="flags/" + file)
    print('downloaded: flags/' + file)
    
```

```python
icons.apply(lambda row : download_img(row['icon_url'], row['icon_file']), axis = 1) 
```
